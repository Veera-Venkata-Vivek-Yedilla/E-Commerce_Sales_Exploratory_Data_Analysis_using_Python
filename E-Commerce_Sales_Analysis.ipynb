### Descriptive Statistics

### 4.1 Univariate Analysis

#### Histogram of Product Prices

```python
sns.histplot(df['price'], bins=20)
ax = plt.title('Distribution of Product Prices')
plt.show()
```

#### Countplot of Review Scores

```python
# First graph: Default settings
sns.countplot(x='review_score', data=df)
plt.title('Distribution of Review Scores')
plt.xlabel('Review Score')
plt.ylabel('Count')
plt.show()

# Second graph: Enhanced visibility with larger figure size, rotated x-axis labels, and count labels
plt.figure(figsize=(18,9))
aaa = sns.countplot(x='review_score', data=df)
plt.title('Distribution of Review Scores')
plt.xlabel('Review Score')
plt.ylabel('Count')
plt.xticks(rotation=45)
for bars in aaa.containers:
    aaa.bar_label(bars)
plt.show()
```

#### Countplot of Categories

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='category', data=df)
plt.title('Distribution of Categories')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

### 4.2 Bivariate Analysis

#### Scatter Plot of Price vs Review Score

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='price', y='review_score', data=df)
plt.title('Price vs Review Score')
plt.xlabel('Price')
plt.ylabel('Review Score')
plt.show()
```

#### Box Plot of Price by Category

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='price', data=df)
plt.title('Price by Category')
plt.xlabel('Category')
plt.ylabel('Price')
plt.show()
```

### 4.3 Multivariate Analysis

#### Pairplot

```python
sns.pairplot(df[['price', 'review_score', 'review_count']])
plt.show()
```

### 4.4 Time Series Analysis

#### Monthly Sales Trend for the Top 5 Products

```python
# Calculate total monthly sales for each product
df['total_sales'] = df[[f'sales_month_{i}' for i in range(1, 13)]].sum(axis=1)
```

```python
# Get top 5 products by total sales
top_5_products = df.nlargest(5, 'total_sales')
```

```python
# Plot Monthly Sales Trends for Top 5 Products
plt.figure(figsize=(12, 6))
for index, row in top_5_products.iterrows():
    plt.plot(range(1, 13), row[[f'sales_month_{i}' for i in range(1, 13)]], label=row['product_name'])
    plt.title('Monthly Sales Trends for Top 5 Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

#### Monthly Sales Trends for Top 5 Categories

```python
# Plot Monthly Sales Trends for Top 5 Categories
plt.figure(figsize=(12, 6))
for index, row in top_5_products.iterrows():
    plt.plot(range(1, 13), row[[f'sales_month_{i}' for i in range(1, 13)]], label=row['category'])
    plt.title('Monthly Sales Trends for Top 5 Categories')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

### 4.5 Grouped Analysis

#### Average Price by Category

```python
avg_price_by_category = df.groupby('category')['price'].mean().reset_index()
plt.figure(figsize=(12, 6))
bx = sns.barplot(y='price', x='category', data=avg_price_by_category)
plt.title('Average Price by Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
for bars in bx.containers:
    bx.bar_label(bars)
plt.show()
```

#### Total Sales by Category

```python
total_sales_by_category = df.groupby('category')['total_sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='total_sales', y='category', data=total_sales_by_category)
plt.title('Total Sales by Category')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.show()
```

#### Calculate the total sales for each month

```python
monthly_sales = df[[f'sales_month_{i}' for i in range(1, 13)]].sum()
```

```python
monthly_sales.index = [f'Month {i}' for i in range(1, 13)]
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=monthly_sales.index, y=monthly_sales)
plt.title('Total Sales for Each Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

#### Total Sales for the Whole Year

```python
df['total_sales'] = df[[f'sales_month_{i}' for i in range(1, 13)]].sum(axis=1)
total_sales_year = df['total_sales'].sum()
total_sales_df = pd.DataFrame({'Total Sales': [total_sales_year]})
plt.figure(figsize=(4,6))
ax = sns.barplot(x=total_sales_df.index,y=total_sales_df['Total Sales'])
plt.title('Total Sales for the Whole Year')
plt.xlabel('')
plt.ylabel('Total Sales')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```
